#### Mike Bossy Stats and Awards

```python
# imports
import pandas as pd
import numpy as np
import zipfile
import os
```

```python
# unzip the folder
with zipfile.ZipFile('data/HockeyData.zip', 'r') as zip_ref:
    zip_ref.extractall('data')
```

**Hall of Fame dataframe**

```python
# read hall of fame csv file
hof = pd.read_csv('data/HOF.csv')
```

```python
# check if Mike Bossy has been admitted to the hall of fame 
hof.loc[hof['name']=='Mike Bossy'] 
```

**Player awards**

```python
# read awards csv file
awards = pd.read_csv('data/AwardsPlayers.csv')
```

```python
# observe the awards given to Bossy
bossy_awards = awards.loc[awards['playerID'] == 'bossymi01'] 
bossy_awards.head()
```
