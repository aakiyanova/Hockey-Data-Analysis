#### Reading and Exploring the Data 

```python
# imports
import pandas as pd
import numpy as np
import zipfile
import os
```

**High-level look at the data**

```python
# unzip the folder
with zipfile.ZipFile('data/HockeyData.zip', 'r') as zip_ref:
    zip_ref.extractall('data')
```

```python
# function to get an overview of the data
def see_file(filepath):
    df = pd.read_csv(filepath, error_bad_lines=False)
    print(df.head())
    print(df.info())
    print(df.describe())
    print(df.shape)
    print(df.isnull().sum())
```
